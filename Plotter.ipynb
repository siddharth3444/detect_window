```python
import numpy as np
d=np.array(mask)
d=d==5
d=(d*255).astype(np.uint8)
print(d.shape)
Image.fromarray(d)
```

```python
import cv2
from google.colab.patches import cv2_imshow

contours, _ = cv2.findContours(d, cv2.RE, cv2.CHAIN_APPROX_NONE)

# Create a copy of the binary mask image to draw contours on
contour_image = cv2.cvtColor(d, cv2.COLOR_GRAY2BGR)

# Draw contours on the image
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# Display the image with contours
cv2_imshow(contour_image)
```

```python
# Approximate contours to polygons
# Approximate contours to polygons
contour_image = cv2.cvtColor(d, cv2.COLOR_GRAY2BGR)
for contour in contours:
  epsilon = 0.05 * cv2.arcLength(contour, True)
  approx_polygon = cv2.approxPolyDP(contour, epsilon, True)

  # Get the vertices of the window polygon
  vertices = approx_polygon.reshape(-1, 2)

  # Create a copy of the binary mask image to draw contours and vertices on


  # Draw contours on the image
  cv2.drawContours(contour_image, [approx_polygon], -1, (0, 255, 0), 2)

  print(vertices)

  # Draw circles at each vertex location
  for vertex in vertices:
      cv2.circle(contour_image, tuple(vertex), 5, (0, 0, 255), -1)

cv2_imshow(contour_image)
```
